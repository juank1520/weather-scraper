```python
import requests
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup


```

```python
def open_url(url):
    try:
        timeout = 5
        driver.implicitly_wait(10) # seconds
        driver.get(url)
        
    except Exception as e:
        print( 'Error obteniendo la url: ', url )
        print(e)
        print('\n')
        print('\n')
    
```

```python
def get_data_from_driver():
    data = {}
    try:
        tablero = driver.find_elements_by_xpath('//div[@class="observation-table ng-star-inserted"]/table/tbody/tr')
        tablero = tablero[0]
        month = tablero.find_element_by_xpath('./td[1]')
        temp = tablero.find_element_by_xpath('./td[2]')
        dew_point = tablero.find_element_by_xpath('./td[3]')
        humidity = tablero.find_element_by_xpath('./td[4]')
        wind_speed = tablero.find_element_by_xpath('./td[5]')
        pressure = tablero.find_element_by_xpath('./td[6]')
        precipitation = tablero.find_element_by_xpath('./td[7]')

        data['day'] = month.text
        data['temp'] = temp.text
        data['dew_point'] = dew_point.text
        data['humidity'] = humidity.text
        data['wind_speed'] = wind_speed.text
        data['pressure'] = pressure.text
        data['precipitation'] = precipitation.text
        return data
    except Exception as e:
        print ('Error obteniendo informaci√≥n de la url: ', driver.current_url)
        print(e)
        print('\n')
        print('\n')
        return None


        
```

```python
def series_to_dataframe( data ):

    df = pd.DataFrame()
    for i, d in enumerate(data):
        splited = pd.Series(data[d]).str.split('\n')
        ndf = pd.DataFrame({ d: splited[0][1:]})
        df = pd.concat([df, ndf], axis=1)

    df[['temp_max','temp_avg', 'temp_min']] = df.temp.str.split(" ",expand=True,)
    df[['dew_point_max','dew_point_avg', 'dew_point_min']] = df.dew_point.str.split(" ",expand=True,)
    df[['humidity_max','humidity_avg', 'humidity_min']] = df.humidity.str.split(" ",expand=True,)
    df[['wind_speed_max','wind_speed_avg', 'wind_speed_min']] = df.wind_speed.str.split(" ",expand=True,)
    df[['pressure_max','pressure_avg', 'pressure_min']] = df.pressure.str.split(" ",expand=True,)

    df = df.drop(columns=['temp', 'dew_point', 'humidity', 'wind_speed', 'pressure'])
    
    return df


                


```

```python
options = webdriver.ChromeOptions()
options.add_argument('--incognito')
driver = webdriver.Chrome(executable_path='../chromedriver', options=options)



url_base = 'https://www.wunderground.com/history/monthly/gt/guatemala-city/MGGT/date/'


data_frames_list = []

for i in range(5):                                   # Years
    for o in range(12):                              # Months
        date = str(2015+i) + '-' + str(o + 1)        # Start in the year: 2015
        new_url = url_base + date
        open_url(new_url)
        data = get_data_from_driver()
        df = series_to_dataframe( data )
        df['month'] = str(o + 1)
        df['year'] = str(2015+i)
        data_frames_list.append(df)
        print(date)

result = pd.concat(data_frames_list, ignore_index=True, keys=None, levels=None, names=None, verify_integrity=False,)
result.to_csv('datos/data.csv')
driver.close()


# urls[0]
# open_url(urls[0])
# data = get_data_from_driver()
# df = series_to_dataframe( data )
```
